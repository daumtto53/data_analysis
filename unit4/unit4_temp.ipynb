```python
import csv
import matplotlib.pyplot as plt

f = open('..\seoul.csv')
data = csv.reader(f)
next(data)
maxresult = []
minresult = []
maxcount = 0
mincount = 0

for row in data :
	if row[-1] != '' :
		maxresult.append(float(row[-1]))
	if row[len(row) - 2] != '' :
		minresult.append(float(row[len(row) - 2]))
plt.plot(maxresult, color = 'r', label = 'max')
plt.show()
plt.plot(minresult, color = 'b', label = 'min')
plt.show()
plt.hist(maxresult, bins = 100, color = 'r', label = 'max')
plt.legend()
plt.show()
plt.hist(minresult, bins = 100, color = 'b', label = 'min')
plt.show()
print('maxcount: ', maxcount, ' ', 'mincount: ', mincount)
```
