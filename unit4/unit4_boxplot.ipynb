```python
import csv

f = open('..\seoul.csv')
data = csv.reader(f)
next(data)
monthtemp = []
for i in range(0,12) :
	monthtemp.append([])
print(len(monthtemp))
for row in data :
	if row[-1] != '' :
		splitdata = row[0].split('-')[1]
		monthtemp[int(splitdata) - 1].append(float(row[-1]))

import matplotlib.pyplot as plt
plt.boxplot(monthtemp)
plt.show()
```
