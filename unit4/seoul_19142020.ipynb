```python
import matplotlib.pyplot as plt
import csv

f1 = open('..\seoul.csv')

data_f1 = csv.reader(f1)

arr_1960 = []
arr_2020 = []

year = -1
month = -1

next(data_f1)
# 평균기온
# for row in data_f1 :
#     year = int(row[0].split('-')[0])
#     month = int(row[0].split('-')[1])
#     if year == 1960 and month >= 6 and month <= 8 :
#         if row[2] == '' :
#             arr_1960.append(float(0))
#         else :
#             arr_1960.append(float(row[2]))
#     if year == 2020 and month >= 6 and month <= 8 :
#         if row[2] == '' :
#             arr_2020.append(float(0))
#         else :
#             arr_2020.append(float(row[2]))

# 최고기온
for row in data_f1 :
    year = int(row[0].split('-')[0])
    month = int(row[0].split('-')[1])
    if year == 1965 and month >= 6 and month <= 8 :
        if row[-1] == '' :
            arr_1960.append(float(0))
        else :
            arr_1960.append(float(row[-1]))
    if year == 2020 and month >= 6 and month <= 8 :
        if row[-1] == '' :
            arr_2020.append(float(0))
        else :
            arr_2020.append(float(row[-1]))

plt.title('seoul temp diff : 1965 vs 2020')
plt.boxplot([arr_1960, arr_2020])
plt.legend()
plt.show()
```
