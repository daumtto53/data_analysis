```python
import csv
import matplotlib.pyplot as plt 

f = open('..\\public_t_time_utf8.csv', 'r', encoding='utf-8')
data = csv.reader(f)
next(data)
next(data)

result = []
res = []

for row in data :
    row[4 :] = map(int, row[4 : ])
    if '판교' in row[3] : 
        print(row)
        res.append
    if row[11] > 120000 :
        res.append(row[4 :])
        print(row)
    result.append(row[11] + row[13] + row[15])
result.sort()


plt.bar(range(len(result)), result)
plt.show()
plt.boxplot(res)
plt.show()
```
