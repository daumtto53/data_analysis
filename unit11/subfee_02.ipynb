```python
#-*coding:utf-8-*-
import csv
import matplotlib.pyplot as plt

f = open('..\\public_t_utf.csv', 'r', encoding='utf-8')
data = csv.reader(f)
next(data)

label = ["유임 승차", "유임 하차", "무임 승차", "무임 하차"]
for row in data :
    for i in range(4, 8) :
        row[i] = int(row[i])
    plt.rc('font', family='Malgun Gothic')
    plt.title(row[3] + ' ' + row[1])
    plt.pie(row[4:8], labels=label, autopct='%.2f%%')
    plt.axis('equal')
    plt.savefig("..\\test\\" + row[3] + '_' + row[1] + '.png')
    plt.show()
```
