```python
import csv
import matplotlib.pyplot as plt

f = open("..\\gender.csv")
data = csv.reader(f)
province = input('enter province')

COUNTY_NUM = 1
SUM_NUM = 2
MAN_NUM = 101
GIRL_NUM = 101

m_pop = []
w_pop = []
gender_pop_diff = [] 

for row in data :
    if province in row[0] :
            for column in row[COUNTY_NUM + SUM_NUM : COUNTY_NUM + MAN_NUM + SUM_NUM] :
                m_pop.append(int(column.replace(',','')))
            for column in row[COUNTY_NUM + MAN_NUM + SUM_NUM * 2 : ] :
                w_pop.append(int(column.replace(',','')))
            break
for i in range(len(w_pop)) :
    gender_pop_diff.append(m_pop[i] - w_pop[i])
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(province + '\'s male - female diff bar graph')
plt.bar(range(len(gender_pop_diff)), gender_pop_diff)
plt.show()
```
