```python
# 제주도 인구

import csv
import matplotlib.pyplot as plt

f = open("..\\gender.csv")
data = csv.reader(f)
province = input('enter province')

COUNTY_NUM = 1
SUM_NUM = 2
MAN_NUM = 101
GIRL_NUM = 101

pie_data = []

for row in data :
    man_n = 0
    woman_n = 0
    if province in row[0] :
        for i in row[COUNTY_NUM + SUM_NUM : COUNTY_NUM + SUM_NUM + MAN_NUM] :
            man_n += int(i.replace(',',''))
        for i in row[COUNTY_NUM + SUM_NUM + MAN_NUM + SUM_NUM : ] :
            woman_n += int(i.replace(',',''))
        pie_data.append(man_n)
        pie_data.append(woman_n)
        break
print(pie_data)


plt.rc('font', family='Malgun Gothic')
plt.axis('equal')
plt.pie(pie_data, labels=['남','여'], autopct="%.2f%%", startangle=90)
plt.title(province + '지역의 남녀 성별 비율')
plt.show()
```
