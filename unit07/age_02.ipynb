```python
import csv
import matplotlib.pyplot as plt

f = open('..\\age.csv')
data = csv.reader(f)

result = []
for row in data :
    if '동탄2동' in row[0] :
        for i in row[3 : ] :
            result.append(int(i.replace(',','')))

plt.rc('font', family='Malgun Gothic')
plt.title('동탄2동 인구연령데이터')
plt.bar(range(101), result)
plt.show()
```
