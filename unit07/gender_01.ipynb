```python
import csv
import matplotlib.pyplot as plt

f = open('..\\gender.csv')
data = csv.reader(f)

COUNTY_NUM = 1
SUM_NUM = 2
MAN_NUM = 101
GIRL_NUM = 101

man_data = []
woman_data = []

for row in data :
    if '2' in row[0] :
        for i in row[COUNTY_NUM + SUM_NUM: COUNTY_NUM + SUM_NUM + MAN_NUM] :
            man_data.append(int(i.replace(',','')) * -1)
        for i in row[COUNTY_NUM + SUM_NUM + MAN_NUM + SUM_NUM : ] :
            woman_data.append(int(i.replace(',','')))

plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('2    ')
plt.barh(range(101), man_data, label='')
plt.barh(range(101), woman_data, label='')
plt.legend()
plt.show()

```
