```python
import matplotlib.pyplot as plt 
import random

plt.scatter([1,2,3,4], [10,30,20,40])
plt.show()
x = [] 
y = [] 
size = []
for i in range(100) :
    x.append(random.randint(50, 100))
    y.append(random.randint(50, 100))
    size.append(random.randint(10, 100))
plt.scatter(x,y,s=size, c=range(100), cmap='jet')
plt.colorbar()
plt.show()
```
