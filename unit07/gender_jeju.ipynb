```python
import csv
import matplotlib.pyplot as plt

f = open("..\\gender.csv")
data = csv.reader(f)
province = input('enter province')

COUNTY_NUM = 1
SUM_NUM = 2
MAN_NUM = 101
GIRL_NUM = 101

m_pop = []
w_pop = []

for row in data :
    if province in row[0] :
            for column in row[COUNTY_NUM + SUM_NUM : COUNTY_NUM + MAN_NUM + SUM_NUM] :
                m_pop.append(int(column.replace(',','')))
            for column in row[COUNTY_NUM + MAN_NUM + SUM_NUM * 2 : ] :
                w_pop.append(int(column.replace(',','')))
            break


plt.plot(m_pop, label='Male')
plt.plot(w_pop, label='Female')
plt.legend()
plt.show()
```
