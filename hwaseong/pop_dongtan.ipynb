```python
import csv
import matplotlib.pyplot as plt 
import numpy as np

f = open('pop.csv')
fd = open('SocialLandArea.csv', encoding='utf-8')
data = csv.reader(f)
land = csv.reader(fd)

set_1 = ['우정', '향남', '남양', '마도', '송산', '서신', '팔탄', '장안', '양감', '새솔']
set_2 = ['비봉', '봉담', '정남', '매송', '기배', '화산', '진안']
set_3 = ['병점1동', '병점2동', '반월', '동탄1동', '동탄2동', '동탄3동']
set_4 = ['동탄4동', '동탄5동', '동탄6동', '동탄7동', '동탄8동']

city_set = [set_1, set_2, set_3, set_4]
header = next(data)

pop_gross_dongtan = 0
pop_gross_etc = 0

pop_gross_list = []
pop_gross_string = ['동탄', '동탄 외 지역']

for row in data :
    for j in range(4) :
        for i in range(len(city_set[j])) :
            if city_set[j][i] in row[0] and '화성' in row[0] :
                if '동탄' in row[0] :
                    pop_gross_dongtan += int(row[1].replace(',',''))
                else :
                    pop_gross_etc += int(row[1].replace(',',''))

plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(15,10))
plt.title("2021년 3월 화성시 동탄, 그 외 지역 인구수 비율")
plt.pie([pop_gross_dongtan, pop_gross_etc], labels=[str(pop_gross_dongtan), str(pop_gross_etc)], autopct='%4.2f%%')
plt.legend(pop_gross_string)
plt.axis('equal')
plt.savefig('.\\g_dongtan_vs_etc.jpg')
plt.show()

plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(15,10))
plt.title("2021년 3월 화성시 동탄 vs 그 외 지역 인구수")
plt.ylabel('인구 수')
plt.xticks(rotation=45)
plt.bar(np.arange(2), [pop_gross_dongtan, pop_gross_etc], tick_label=pop_gross_string, width=0.5)
plt.savefig('.\\g_dongtan_vs_etc_pop.jpg')
plt.show()
```
